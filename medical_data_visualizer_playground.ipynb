```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
examinations = pd.read_csv('./medical_examination.csv')
```

```python
medical_diagnosis = examinations.copy()
medical_diagnosis['bmi'] = medical_diagnosis['weight'] / ((medical_diagnosis['height'] / 100) **2)
```

```python
medical_diagnosis['overweight'] = np.where(medical_diagnosis['bmi'] > 25, 1, 0)
```

```python
medical_diagnosis.head()
```

```python
medical_diagnosis['cholesterol'] = np.where(medical_diagnosis['cholesterol'] > 1, 1, 0)
medical_diagnosis['gluc'] = np.where(medical_diagnosis['gluc'] > 1, 1, 0)
```

```python
medical_diagnosis.head()
medical_diagnosis['gluc'].value_counts()
```

```python
health_indicators = ['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke']
medical_diagnosis_long = medical_diagnosis.melt(id_vars='cardio', value_vars=health_indicators, var_name='indicator').value_counts().to_frame()
medical_diagnosis_long = medical_diagnosis_long.reset_index(level=['cardio', 'indicator', 'value'])
medical_diagnosis_long.rename(columns={'count':'total'}, inplace=True)
cat_grid = sns.catplot(medical_diagnosis_long, x='indicator', y='total', hue='value', col='cardio', order=health_indicators, kind='bar')
cat_figure = cat_grid.fig
cat_figure.savefig('catplot_jupyther.png')
```

```python
medical_diagnosis_clean = medical_diagnosis.copy()
valid_diastolic_values = medical_diagnosis['ap_lo'] <= medical_diagnosis['ap_hi']
height_big_enough = medical_diagnosis['height'] >= medical_diagnosis['height'].quantile(0.025)
height_not_to_big = medical_diagnosis['height'] <= medical_diagnosis['height'].quantile(0.975)
weight_enough = medical_diagnosis['weight'] >= medical_diagnosis['weight'].quantile(0.025)
weight_not_to_fat = medical_diagnosis['weight'] <= medical_diagnosis['weight'].quantile(0.975)

medical_diagnosis_clean = medical_diagnosis.loc[(valid_diastolic_values\
                                                 & height_big_enough\
                                                 & height_not_to_big\
                                                 & weight_enough\
                                                 & weight_not_to_fat)]
medical_diagnosis_clean.head()
```

```python
columns_of_interest = ['id', 'age', 'sex', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol',
       'gluc', 'smoke', 'alco', 'active', 'cardio','overweight']

medical_diagnosis_clean_corr = medical_diagnosis_clean[columns_of_interest].corr()
only_show_lower_triangle = np.triu(medical_diagnosis_clean_corr)
plt.figure(figsize=(12, 12))
sns.heatmap(medical_diagnosis_clean_corr, annot=True, fmt='.1f', cbar=False, mask=only_show_lower_triangle)
```
